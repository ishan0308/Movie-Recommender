```python
import pandas as pd 
```

```python
tmbd_movies = pd.read_csv('data/tmdb_5000_movies.csv')
tmbd_credits = pd.read_csv('data/tmdb_5000_credits.csv') 
```

```python
print(tmbd_movies.columns)
tmbd_movies.head(2)
```

```python
print(tmbd_credits.columns)
tmbd_credits.head(2)
```

```python
movies = tmbd_movies.merge(tmbd_credits,on='title')
```

```python
print(movies.columns)
movies.head(2)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(2)
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
ex_data = str(movies.iloc[0]['genres'])
print(ex_data)
for data in ast.literal_eval(ex_data):
        print(data, type(data), data['name']) 
```

```python
def convert_data(cell):
    reformed_data = []
    for data in ast.literal_eval(cell):
        reformed_data.append(data['name']) 
    return reformed_data 
```

```python
movies['genres'] = movies['genres'].apply(convert_data)
movies.head(2)
```

```python
movies['keywords'] = movies['keywords'].apply(convert_data)
movies['cast'] = movies['cast'].apply(convert_data)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
movies.head(2)
```

```python
def fetch_director(cell):
    reformed_list = []
    for data in ast.literal_eval(cell):
        if data['job'] == 'Director':
            reformed_list.append(data['name'])
    return reformed_list 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
def collapse(cell):
    reformed_list = []
    for data in cell:
        reformed_list.append(data.replace(" ",""))
    return reformed_list
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head(2)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
new_movies = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new_movies.head(2)
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
new_movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
x = new_movies[new_movies['title'] == 'The Lego Movie'].index[0]
print(x)
list(enumerate(similarity[x]))
```

```python
def recommend(movie):
    index = new_movies[new_movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
